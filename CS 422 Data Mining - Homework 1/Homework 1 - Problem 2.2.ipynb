```python
import pandas as pd
import numpy as np 
import scipy
from sklearn.preprocessing import Imputer

path = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

df1 = df2 = df3 = pd.read_csv(
                    path, delim_whitespace = True, header= None, na_values=['?'],
                    names = ['mpg', 'cylinders', 'displacement','horsepower','weight', 'acceleration', 'model_year', 'origin', 'name']
                )

print("DEFAULT DATAFRAME")
df1.describe()
```

```python
print("ZERO FILL DATAFRAME")
df2 = df2.fillna(0)
df2.describe()
```

```python
print("MEAN NaN IMPUTED")
imp = Imputer(missing_values = "NaN", strategy = "mean", axis=0)
df3["horsepower"] = imp.fit_transform(df3[["horsepower"]])
df3.describe()
```
