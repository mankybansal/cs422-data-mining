```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
from sklearn.cluster import KMeans

from sklearn import preprocessing
from sklearn import metrics
```

```python
dataset = load_boston()
df = pd.DataFrame(data = dataset.data, columns = dataset.feature_names)
```

```python
clustModel2 = KMeans(n_clusters=2, init='k-means++')
clustModel3 = KMeans(n_clusters=3, init='k-means++')
clustModel4 = KMeans(n_clusters=4, init='k-means++')
clustModel5 = KMeans(n_clusters=5, init='k-means++')
clustModel6 = KMeans(n_clusters=6, init='k-means++')

clustLabels2 = clustModel2.fit_predict(df)
clustLabels3 = clustModel3.fit_predict(df)
clustLabels4 = clustModel4.fit_predict(df)
clustLabels5 = clustModel5.fit_predict(df)
clustLabels6 = clustModel6.fit_predict(df)

silhouetteAvg2 = metrics.silhouette_score(df, clustLabels2)
silhouetteAvg3 = metrics.silhouette_score(df, clustLabels3)
silhouetteAvg4 = metrics.silhouette_score(df, clustLabels4)
silhouetteAvg5 = metrics.silhouette_score(df, clustLabels5)
silhouetteAvg6 = metrics.silhouette_score(df, clustLabels6)

print("Silhouette Average Model 2: ", silhouetteAvg2)
print("Silhouette Average Model 3: ", silhouetteAvg3)
print("Silhouette Average Model 4: ", silhouetteAvg4)
print("Silhouette Average Model 5: ", silhouetteAvg5)
print("Silhouette Average Model 6: ", silhouetteAvg6)
```

```python
targetValues = dataset.target
targetLabels = pd.qcut(x = targetValues, q = 3, labels=[1,2,3])
classLabels = targetLabels.astype('int32') - 1
print(classLabels)
```

```python
homogeneity = metrics.homogeneity_score(classLabels, clustLabels3)
completeness = metrics.completeness_score(classLabels, clustLabels3)
print("Homogeneity: ", homogeneity)
print("Completeness: ", completeness)
```

```python
df['MEDV'] = dataset.target
df['CLUST'] = clustLabels3
df['CLASS'] = classLabels
```
